```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV, cross_val_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import warnings
import joblib
import glob
from datetime import timedelta
import time
from tqdm import tqdm
import pickle

# Configure TensorFlow to properly detect and use GPU if available
# Do not disable CUDA devices by default
if 'CUDA_VISIBLE_DEVICES' in os.environ:
    print(f"CUDA_VISIBLE_DEVICES is already set to: {os.environ['CUDA_VISIBLE_DEVICES']}")
else:
    # Let TensorFlow decide which devices to use
    print("Allowing TensorFlow to detect available devices")

# Reduce TensorFlow logging
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"
os.environ["TF_ENABLE_ONEDNN_OPTS"] = "0"  # Disable oneDNN optimizations which can cause issues

# TensorFlow and Keras setup with error handling and GPU detection
try:
    import tensorflow as tf
    # IMPORTANT: Keep eager execution enabled for Jupyter compatibility
    print(f"TensorFlow version: {tf.__version__}")
    
    # Check for GPU availability and show device information
    gpus = tf.config.list_physical_devices('GPU')
    if gpus:
        print(f"Found {len(gpus)} GPU(s):")
        for gpu in gpus:
            print(f"  {gpu.name}")
        
        # Configure TensorFlow to use memory growth to avoid allocating all GPU memory at once
        for gpu in gpus:
            try:
                tf.config.experimental.set_memory_growth(gpu, True)
                print(f"  Memory growth enabled for {gpu.name}")
            except RuntimeError as e:
                print(f"  Error setting memory growth for {gpu.name}: {e}")
    else:
        print("No GPU found. Using CPU for TensorFlow operations.")
    
    # Import Keras components
    from tensorflow.keras.models import Sequential, load_model, save_model
    from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional
    from tensorflow.keras.optimizers import Adam
    from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
    from tensorflow.keras import regularizers
    
    # Set random seeds for reproducibility
    tf.random.set_seed(42)
    
    HAS_TENSORFLOW = True
    print("TensorFlow and Keras imported successfully")
    
except ImportError:
    print("TensorFlow or Keras not available. LSTM models will be skipped.")
    HAS_TENSORFLOW = False
except Exception as e:
    print(f"Error initializing TensorFlow/Keras: {str(e)}")
    print("LSTM models will be skipped.")
    HAS_TENSORFLOW = False

# Set random seeds for reproducibility
np.random.seed(42)
warnings.filterwarnings('ignore')

class StockSelectionModel:
    def __init__(self, data_dir="processed_data/", results_dir="model_results_v4/", lookback=5, use_lstm=True, 
                 risk_free_rate=0.04, trading_days_per_year=252):
        """
        Initialize the stock selection model.
        
        Parameters:
        -----------
        data_dir : str
            Directory containing processed CSV files
        results_dir : str
            Directory to save model results
        lookback : int
            Number of days to look back for feature engineering
        use_lstm : bool
            Whether to use LSTM models
        risk_free_rate : float
            Annual risk-free rate for Sharpe ratio calculation
        trading_days_per_year : int
            Number of trading days in a year for annualizing returns
        """
        self.data_dir = data_dir
        self.results_dir = results_dir
        self.lookback = lookback
        self.use_lstm = use_lstm and HAS_TENSORFLOW
        self.ticker_data = {}
        self.models = {}
        self.feature_sets = {}
        self.results = {}
        self.scalers = {}  # Keep track of scalers for each ticker and feature set
        self.feature_names = {}  # Keep track of feature names for each ticker and feature set
        self.best_params = {}  # Store best hyperparameters
        self.benchmark_metrics = {}  # Store benchmark metrics for comparison
        self.risk_free_rate = risk_free_rate
        self.trading_days_per_year = trading_days_per_year
        
        # Create results directory if it doesn't exist
        if not os.path.exists(results_dir):
            os.makedirs(results_dir)
    
    def load_data(self, file_pattern="*_all_data.csv"):
        """
        Load data for all tickers matching the file pattern.
        
        Parameters:
        -----------
        file_pattern : str
            Pattern to match CSV files
        """
        file_paths = glob.glob(os.path.join(self.data_dir, file_pattern))
        
        for file_path in file_paths:
            ticker = os.path.basename(file_path).split('_')[0]
            data = pd.read_csv(file_path)
            data['Date'] = pd.to_datetime(data['Date'])
            data = data.sort_values('Date').reset_index(drop=True)
            
            # Same for PE and ROE values that change infrequently
            fin_cols = ['weighted_PE_Ratio', 'weighted_ROE']
            for col in fin_cols:
                if col in data.columns and data[col].nunique() < 10:
                    # Forward fill to handle any missing values
                    data[col] = data[col].ffill().bfill()
            
            # Make sure we're using Adjusted_close when available for consistency
            if 'Adjusted_close' in data.columns and 'Close' in data.columns:
                # Verify Adjusted_close exists and is non-null before using it
                if data['Adjusted_close'].notna().all():
                    # Calculate log returns based on Adjusted_close if not already present
                    if 'log_return' not in data.columns:
                        data['log_return'] = np.log(data['Adjusted_close'] / data['Adjusted_close'].shift(1))
                    
                    # Recalculate next_day_return based on Adjusted_close if not present or if inconsistent
                    if 'next_day_return' not in data.columns:
                        data['next_day_return'] = data['Adjusted_close'].pct_change(1).shift(-1)
                    else:
                        # Calculate what next_day_return should be based on Adjusted_close
                        expected_next_day_return = data['Adjusted_close'].pct_change(1).shift(-1)
                        
                        # Check if next_day_return needs recalculation
                        inconsistent_values = (data['next_day_return'] - expected_next_day_return).abs() > 0.0001
                        if inconsistent_values.any():
                            print(f"Recalculating next_day_return for {ticker} based on Adjusted_close "
                                 f"({inconsistent_values.sum()} inconsistent values found)")
                            data['next_day_return'] = expected_next_day_return
                else:
                    print(f"Warning: Adjusted_close contains NaN values for {ticker}. Using Close instead.")
                    # Use Close if Adjusted_close has NaN values
                    if 'log_return' not in data.columns:
                        data['log_return'] = np.log(data['Close'] / data['Close'].shift(1))
                    
                    if 'next_day_return' not in data.columns:
                        data['next_day_return'] = data['Close'].pct_change(1).shift(-1)
            
            self.ticker_data[ticker] = data
            print(f"Loaded data for {ticker}: {len(data)} rows")
    
    def create_feature_sets(self):
        """
        Create different feature sets for comparison.
        """
        # Get a sample ticker to check column names
        sample_ticker = list(self.ticker_data.keys())[0]
        columns = self.ticker_data[sample_ticker].columns
        
        self.feature_sets = {
            "price_only": ['log_return'],  # Simplified to just log_return
            "financial_only": [col for col in columns if col in ['weighted_PE_Ratio', 'weighted_ROE']],
            "reddit_sentiment": [col for col in columns if col.startswith('reddit_')],
            "news_sentiment": [col for col in columns if col.startswith('news_')],
            "mda_sentiment": [col for col in columns if col.startswith('mda_')],
        }
        
        # Add all sentiment features
        all_sentiment = []
        for features in ['reddit_sentiment', 'news_sentiment', 'mda_sentiment']:
            all_sentiment.extend(self.feature_sets[features])
        self.feature_sets["all_sentiment"] = all_sentiment
        
        # Add combined feature sets
        self.feature_sets["financial_and_sentiment"] = (
            self.feature_sets["financial_only"] + self.feature_sets["all_sentiment"]
        )
        self.feature_sets["price_and_sentiment"] = (
            self.feature_sets["price_only"] + self.feature_sets["all_sentiment"]
        )
        self.feature_sets["price_and_financial"] = (
            self.feature_sets["price_only"] + self.feature_sets["financial_only"]
        )
        self.feature_sets["all_features"] = (
            self.feature_sets["price_only"] + self.feature_sets["financial_only"] + 
            self.feature_sets["all_sentiment"]
        )
        
        print(f"Created {len(self.feature_sets)} feature sets")
        for name, features in self.feature_sets.items():
            print(f"  {name}: {len(features)} features")
    
    def engineer_features(self, data, feature_set=None):
        """
        Engineer features from the raw data to avoid look-ahead bias.
        
        Parameters:
        -----------
        data : pandas.DataFrame
            Raw data for a ticker
        feature_set : list, optional
            List of base features to use for engineering. If None, uses all features.
            
        Returns:
        --------
        pandas.DataFrame
            Data with engineered features
        """
        df = data.copy()
        
        # If feature_set is provided, filter base features
        if feature_set is not None:
            # Always keep Date, ticker, and next_day_return columns
            essential_cols = ['Date', 'ticker', 'next_day_return']
            feature_cols = [col for col in df.columns if col in feature_set or col in essential_cols]
            df = df[feature_cols]
        
        # Add lagged features for price data (log_return)
        price_cols = ['log_return']
        for col in price_cols:
            if col in df.columns:
                for lag in range(1, self.lookback + 1):
                    df[f'{col}_lag_{lag}'] = df[col].shift(lag)
        
        # Add rolling means for log_return
        for col in price_cols:
            if col in df.columns:
                for window in [5, 10, 20]:
                    df[f'{col}_rolling_{window}'] = df[col].rolling(window=window).mean()
        
        # Add day of week and month features
        df['day_of_week'] = df['Date'].dt.dayofweek
        df['month'] = df['Date'].dt.month
        df['quarter'] = df['Date'].dt.quarter
        
        # Ensure we don't have look-ahead bias by shifting sentiment features
        sentiment_cols = [col for col in df.columns if col.startswith(('reddit_', 'news_', 'mda_'))]
        for col in sentiment_cols:
            # Shift by 1 to ensure we only use data available at prediction time
            df[col] = df[col].shift(1)
        
        # Drop NaN values created by lag and rolling operations
        df = df.dropna().reset_index(drop=True)
        
        return df
    
    def prepare_data_for_model(self, data, feature_set, ticker, set_name, target='next_day_return', test_size=0.05):
        """
        Prepare data for model training and testing.
        
        Parameters:
        -----------
        data : pandas.DataFrame
            Data with engineered features
        feature_set : list
            List of features to use for modeling
        ticker : str
            Ticker symbol for identification
        set_name : str
            Name of the feature set for tracking
        target : str
            Target variable to predict
        test_size : float
            Proportion of data to use for testing
            
        Returns:
        --------
        tuple
            (X_train, X_test, y_train, y_test, scaler, feature_names, train_data, test_data)
        """
        # Make sure we have the Date column for reference
        if 'Date' not in data.columns:
            raise ValueError("Data must contain a 'Date' column")
        
        # Ensure all requested features exist in the dataframe
        valid_features = [f for f in feature_set if f in data.columns]
        
        # Add engineered features that might be based on the requested features
        engineered_features = []
        for feature in feature_set:
            lagged_features = [col for col in data.columns if col.startswith(f"{feature}_lag_")]
            rolling_features = [col for col in data.columns if col.startswith(f"{feature}_rolling_")]
            engineered_features.extend(lagged_features + rolling_features)
        
        # Add temporal features
        temporal_features = [col for col in data.columns if col in [
            'day_of_week', 'month', 'quarter'
        ]]
        
        # Combine all features
        all_features = valid_features + engineered_features + temporal_features
        
        # Make sure we have unique features
        all_features = list(set(all_features))
        
        # Split data chronologically
        train_size = int(len(data) * (1 - test_size))
        train_data = data.iloc[:train_size]
        test_data = data.iloc[train_size:]
        
        # Scale features
        scaler = StandardScaler()
        X_train = scaler.fit_transform(train_data[all_features])
        X_test = scaler.transform(test_data[all_features])
        
        y_train = train_data[target].values
        y_test = test_data[target].values
        
        # Store scaler and feature names for later use
        key = f"{ticker}_{set_name}"
        self.scalers[key] = scaler
        self.feature_names[key] = all_features
        
        return X_train, X_test, y_train, y_test, scaler, all_features, train_data, test_data
    
    def prepare_lstm_data(self, X, y, lookback):
        """
        Reshape data for LSTM model.
        
        Parameters:
        -----------
        X : numpy.ndarray
            Feature matrix
        y : numpy.ndarray
            Target vector
        lookback : int
            Number of timesteps to look back
            
        Returns:
        --------
        tuple
            (X_lstm, y_lstm)
        """
        X_lstm, y_lstm = [], []
        for i in range(len(X) - lookback):
            X_lstm.append(X[i:i+lookback])
            y_lstm.append(y[i+lookback])
        
        return np.array(X_lstm), np.array(y_lstm)
    
    def build_models(self):
        """
        Build all models for comparison with fixed hyperparameters.
        """
        # Start with standard models that don't require TensorFlow
        self.models = {
            "OLS": LinearRegression(),
            "Ridge": Ridge(alpha=5.0),  # Fixed 
            "Lasso": Lasso(alpha=0.1),  # Fixed 
            "KNN": KNeighborsRegressor(n_neighbors=5),  # Will be optimized later
            "RandomForest": RandomForestRegressor(n_estimators=100, random_state=42),  # Will be optimized later
        }
        
        # Add LSTM only if TensorFlow is available and use_lstm is True
        if self.use_lstm and HAS_TENSORFLOW:
            self.models["LSTM"] = None  # Will be built separately for each dataset
        
        print(f"Built {len(self.models)} models for comparison")
        if "LSTM" not in self.models and self.use_lstm:
            print("Note: LSTM models will be skipped due to TensorFlow configuration")
    
    def optimize_knn(self, X_train, y_train, ticker, feature_set):
        """
        Optimize KNN hyperparameters using cross-validation.
        
        Parameters:
        -----------
        X_train : numpy.ndarray
            Training features
        y_train : numpy.ndarray
            Training target
        ticker : str
            Ticker symbol for identification
        feature_set : str
            Feature set name for identification
            
        Returns:
        --------
        int
            Optimal number of neighbors
        """
        print(f"    Optimizing KNN hyperparameters for {ticker} with {feature_set}...")
        
        # Use a different seed for each ticker/feature set to ensure diversity
        seed = hash(f"{ticker}_{feature_set}") % 10000
        np.random.seed(seed)
        
        # Define range of neighbors to search - vary based on dataset size
        min_n = 2
        max_n = min(100, X_train.shape[0] // 5)  # Limit based on dataset size
        
        # Create a more specific range based on dataset characteristics
        data_size = X_train.shape[0]
        if data_size > 1000:
            # For large datasets, use more neighbors and wider spacing
            n_neighbors_range = list(range(min_n, 30, 3)) + list(range(30, max_n, 5))
        elif data_size > 500:
            # For medium datasets
            n_neighbors_range = list(range(min_n, 20, 2)) + list(range(20, max_n, 3))
        else:
            # For smaller datasets, use more fine-grained search
            n_neighbors_range = list(range(min_n, max_n, 1))
        
        # Set up time series cross-validation (more splits for larger datasets)
        n_splits = max(3, min(10, data_size // 100))
        tscv = TimeSeriesSplit(n_splits=n_splits)
        
        # Store results
        cv_scores = []
        
        # Perform cross-validation for each value of n_neighbors
        for n_neighbors in n_neighbors_range:
            model = KNeighborsRegressor(n_neighbors=n_neighbors, 
                                        weights='uniform',  # Try different weights later
                                        algorithm='auto')
            scores = cross_val_score(model, X_train, y_train, cv=tscv, scoring='neg_mean_squared_error')
            cv_scores.append(np.mean(scores))
        
        # Find the best n_neighbors
        optimal_n = n_neighbors_range[np.argmax(cv_scores)]
        
        # Now try different weights with the optimal n_neighbors
        weights_options = ['uniform', 'distance']
        weights_scores = []
        
        for weights in weights_options:
            model = KNeighborsRegressor(n_neighbors=optimal_n, weights=weights, algorithm='auto')
            scores = cross_val_score(model, X_train, y_train, cv=tscv, scoring='neg_mean_squared_error')
            weights_scores.append(np.mean(scores))
        
        # Find the best weights
        optimal_weights = weights_options[np.argmax(weights_scores)]
        
        print(f"    Optimal n_neighbors for KNN: {optimal_n}, weights: {optimal_weights}")
        
        # Return the best parameters
        return {
            'n_neighbors': optimal_n,
            'weights': optimal_weights,
            'algorithm': 'auto'
        }
    
    def optimize_random_forest(self, X_train, y_train, ticker, feature_set):
        """
        Optimize Random Forest hyperparameters using grid search.
        
        Parameters:
        -----------
        X_train : numpy.ndarray
            Training features
        y_train : numpy.ndarray
            Training target
        ticker : str
            Ticker symbol for identification
        feature_set : str
            Feature set name for identification
            
        Returns:
        --------
        dict
            Best hyperparameters
        """
        print(f"    Optimizing Random Forest hyperparameters for {ticker} with {feature_set}...")
        
        # Use a different seed for each ticker/feature set to ensure diversity
        seed = hash(f"{ticker}_{feature_set}") % 10000
        
        # Adjust parameter grid based on dataset size
        data_size = X_train.shape[0]
        n_features = X_train.shape[1]
        
        if data_size > 1000:
            # For larger datasets, use more estimators and deeper trees
            n_estimators = [50, 100, 200, 300]
            max_depth = [None, 10, 20, 30, 40]
            min_samples_split = [2, 5, 10]
        elif data_size > 500:
            # For medium datasets
            n_estimators = [50, 100, 150, 200]
            max_depth = [None, 5, 10, 15, 20]
            min_samples_split = [2, 5, 7]
        else:
            # For smaller datasets, use fewer estimators to avoid overfitting
            n_estimators = [30, 50, 100, 150]
            max_depth = [None, 3, 5, 10, 15]
            min_samples_split = [2, 3, 5]
        
        # Define parameter grid - FIXED: Replaced 'auto' with 1.0 to avoid warning
        param_grid = {
            'n_estimators': n_estimators,
            'max_depth': max_depth,
            'min_samples_split': min_samples_split,
            'max_features': [1.0, 'sqrt', 'log2'],  # Changed 'auto' to 1.0
            'random_state': [seed]
        }
        
        # Set up time series cross-validation (more splits for larger datasets)
        n_splits = max(3, min(5, data_size // 200))
        tscv = TimeSeriesSplit(n_splits=n_splits)
        
        # Create grid search object
        grid_search = GridSearchCV(
            RandomForestRegressor(),
            param_grid,
            cv=tscv,
            scoring='neg_mean_squared_error',
            n_jobs=-1  # Use all available CPU cores
        )
        
        # Perform grid search
        grid_search.fit(X_train, y_train)
        
        # Get best parameters
        best_params = grid_search.best_params_
        
        print(f"    Best Random Forest parameters for {ticker}/{feature_set}: {best_params}")
        
        return best_params
    
    def build_lstm_model(self, input_shape, ticker=None, feature_set=None):
        """
        Build LSTM model with the given input shape using fixed hyperparameters.
        
        Parameters:
        -----------
        input_shape : tuple
            Shape of input data (lookback, features)
        ticker : str, optional
            Ticker symbol for identification
        feature_set : str, optional
            Feature set name for identification
            
        Returns:
        --------
        keras.models.Sequential
            LSTM model
        """
        try:
            # Set different random seed for each ticker/feature set combination
            if ticker and feature_set:
                seed = hash(f"{ticker}_{feature_set}") % 10000
                tf.random.set_seed(seed)
            else:
                tf.random.set_seed(42)
            
            # Build a simpler LSTM model as benchmark to avoid memory issues
            model = Sequential()
            
            # Vary model architecture based on data complexity
            if input_shape[1] > 30:  # More complex for high-dimensional data
                units = 96
                dropout = 0.3
            elif input_shape[1] > 15:  # Medium complexity
                units = 64
                dropout = 0.2
            else:  # Simpler for low-dimensional data
                units = 32
                dropout = 0.1
            
            # Add LSTM layer with conservative defaults to avoid GPU issues
            model.add(LSTM(units, 
                          input_shape=input_shape, 
                          activation='tanh',
                          return_sequences=False,
                          recurrent_dropout=0.0,  # Explicitly set to 0.0 to avoid GPU issues
                          implementation=1))  # Use implementation=1 which is more memory efficient
            
            model.add(Dropout(dropout))
            model.add(Dense(units//2, activation='relu'))
            model.add(Dropout(dropout/2))
            model.add(Dense(1))
            
            # Use a reasonable optimizer with specified learning rate
            optimizer = Adam(learning_rate=0.001, clipnorm=1.0)  # Add gradient clipping
            model.compile(optimizer=optimizer, loss='mse')
            
            print(f"    Standard LSTM model created successfully for {ticker}/{feature_set}.")
            return model
        except Exception as e:
            print(f"    Error building LSTM model: {str(e)}")
            print("    Attempting to use simpler LSTM configuration...")
            
            # Try with an even simpler LSTM configuration
            try:
                if ticker and feature_set:
                    seed = hash(f"{ticker}_{feature_set}") % 10000
                    tf.random.set_seed(seed)
                else:
                    tf.random.set_seed(42)
                
                # Minimal LSTM model
                model = Sequential()
                model.add(LSTM(32, 
                              input_shape=input_shape, 
                              activation='tanh',
                              recurrent_activation='sigmoid',  # Explicitly set recurrent activation
                              return_sequences=False))
                model.add(Dense(1))
                
                optimizer = Adam(learning_rate=0.001)
                model.compile(optimizer=optimizer, loss='mse')
                
                print(f"    Simplified LSTM model created successfully for {ticker}/{feature_set}.")
                return model
            except Exception as e2:
                print(f"    Error building simplified LSTM model: {str(e2)}")
                print("    Attempting to build fallback Dense model instead...")
                
                # Fallback to a simple Dense model that takes flattened input
                try:
                    if ticker and feature_set:
                        seed = hash(f"{ticker}_{feature_set}") % 10000
                        tf.random.set_seed(seed)
                    else:
                        tf.random.set_seed(42)
                    
                    # Calculate flattened input size
                    flattened_size = input_shape[0] * input_shape[1]
                    
                    # Build a simple dense model
                    model = Sequential()
                    model.add(Dense(64, activation='relu', input_shape=(flattened_size,)))
                    model.add(Dropout(0.2))
                    model.add(Dense(32, activation='relu'))
                    model.add(Dropout(0.1))
                    model.add(Dense(1))
                    
                    optimizer = Adam(learning_rate=0.001)
                    model.compile(optimizer=optimizer, loss='mse')
                    
                    # Mark this as a fallback model so we know to reshape inputs
                    model.is_fallback = True
                    
                    print(f"    Fallback Dense model created successfully for {ticker}/{feature_set}.")
                    return model
                except Exception as nested_e:
                    print(f"    Error building fallback model: {str(nested_e)}")
                    return None
    
    def calculate_sharpe_ratio(self, returns, risk_free_rate=None, trading_days=None):
        """
        Calculate the Sharpe ratio for a series of returns.
        
        Parameters:
        -----------
        returns : numpy.ndarray
            Array of returns
        risk_free_rate : float, optional
            Annual risk-free rate (defaults to class value)
        trading_days : int, optional
            Number of trading days in a year (defaults to class value)
            
        Returns:
        --------
        float
            Sharpe ratio
        """
        # Use class defaults if not specified
        if risk_free_rate is None:
            risk_free_rate = self.risk_free_rate
        
        if trading_days is None:
            trading_days = self.trading_days_per_year
        
        # Convert annual risk-free rate to match the frequency of returns
        daily_risk_free = risk_free_rate / trading_days
        
        # Calculate excess returns
        excess_returns = returns - daily_risk_free
        
        # Calculate Sharpe ratio (annualized)
        if np.std(excess_returns) == 0:
            return 0  # Avoid division by zero
        
        sharpe = np.mean(excess_returns) / np.std(excess_returns) * np.sqrt(trading_days)
        
        return sharpe
    
    def calculate_benchmark_metrics(self, data, test_data, test_dates):
        """
        Calculate benchmark metrics for comparison purposes.
        
        Parameters:
        -----------
        data : pandas.DataFrame
            Full dataset for the ticker
        test_data : pandas.DataFrame
            Test portion of the dataset
        test_dates : numpy.ndarray
            Array of dates for the test portion
            
        Returns:
        --------
        dict
            Dictionary containing benchmark metrics
        """
        # Create date index for alignment
        date_index = pd.DatetimeIndex(test_dates)
        
        # Historical mean return (unconditional expectation)
        historical_mean = data['next_day_return'].mean()
        
        # Create a prediction of historical mean for every test point
        y_pred_hist_mean = np.full(len(test_data), historical_mean)
        
        # Random walk prediction (today's return predicts tomorrow)
        # Get the log_return column from test_data
        if 'log_return' in test_data.columns:
            y_pred_random_walk = test_data['log_return'].values
        else:
            # If log_return isn't available, use a lag of the target
            y_pred_random_walk = np.roll(test_data['next_day_return'].values, 1)
            y_pred_random_walk[0] = historical_mean  # Fill first value
        
        # Moving average prediction (5-day)
        if 'log_return' in data.columns:
            # Create a full series of log_return
            full_returns = data['log_return'].copy()
            
            # Calculate 5-day moving average
            ma5 = full_returns.rolling(window=5).mean()
            
            # Extract values for test period
            test_indices = data[data['Date'].isin(test_dates)].index
            y_pred_ma5 = ma5.iloc[test_indices].values
            
            # Fill NaN values with historical mean
            y_pred_ma5 = np.nan_to_num(y_pred_ma5, nan=historical_mean)
        else:
            # Fallback if log_return isn't available
            y_pred_ma5 = y_pred_hist_mean
        
        # Actual values
        y_actual = test_data['next_day_return'].values
        
        # Calculate R² for each benchmark
        r2_hist_mean = r2_score(y_actual, y_pred_hist_mean)
        r2_random_walk = r2_score(y_actual, y_pred_random_walk)
        r2_ma5 = r2_score(y_actual, y_pred_ma5)
        
        # Calculate Sharpe ratios
        sharpe_hist_mean = self.calculate_sharpe_ratio(y_pred_hist_mean)
        sharpe_random_walk = self.calculate_sharpe_ratio(y_pred_random_walk)
        sharpe_ma5 = self.calculate_sharpe_ratio(y_pred_ma5)
        
        # Return all benchmark metrics
        return {
            'historical_mean_r2': r2_hist_mean,
            'random_walk_r2': r2_random_walk,
            'moving_avg_r2': r2_ma5,
            'historical_mean_sharpe': sharpe_hist_mean,
            'random_walk_sharpe': sharpe_random_walk,
            'moving_avg_sharpe': sharpe_ma5
        }
    
    def calculate_metrics(self, y_train, y_pred_train, y_test, y_pred_test):
        """
        Calculate only R-square and directional accuracy metrics.
        
        Parameters:
        -----------
        y_train : numpy.ndarray
            Actual training values
        y_pred_train : numpy.ndarray
            Predicted training values
        y_test : numpy.ndarray
            Actual test values
        y_pred_test : numpy.ndarray
            Predicted test values
            
        Returns:
        --------
        dict
            Dictionary of metrics
        """
        # Calculate R² (R-square)
        train_r2 = r2_score(y_train, y_pred_train)
        test_r2 = r2_score(y_test, y_pred_test)
        
        # Calculate directional accuracy
        train_sign_match = (np.sign(y_pred_train) == np.sign(y_train))
        test_sign_match = (np.sign(y_pred_test) == np.sign(y_test))
        
        train_dir_accuracy = np.mean(train_sign_match)
        test_dir_accuracy = np.mean(test_sign_match)
        
        return {
            'train_r2': train_r2,
            'test_r2': test_r2,
            'train_dir_accuracy': train_dir_accuracy,
            'test_dir_accuracy': test_dir_accuracy
        }
    
    def train_and_evaluate(self, ticker):
        """
        Train and evaluate all models for a given ticker.
        
        Parameters:
        -----------
        ticker : str
            Stock ticker symbol
        """
        print(f"\nTraining and evaluating models for {ticker}")
        
        # Get data for this ticker
        data = self.ticker_data[ticker]
        
        # Initialize results for this ticker
        self.results[ticker] = {}
        self.best_params[ticker] = {}
        self.benchmark_metrics[ticker] = {}
        
        # Loop through each feature set
        for feature_set_name, features in self.feature_sets.items():
            print(f"  Feature set: {feature_set_name}")
            
            # Engineer features specifically for this feature set
            engineered_data = self.engineer_features(data, features)
            
            # Prepare data for this feature set
            X_train, X_test, y_train, y_test, scaler, feature_names, train_data, test_data = self.prepare_data_for_model(
                engineered_data, features, ticker, feature_set_name
            )
            
            # Calculate benchmark metrics for this ticker and feature set
            benchmark_metrics = self.calculate_benchmark_metrics(data, test_data, test_data['Date'].values)
            self.benchmark_metrics[ticker][feature_set_name] = benchmark_metrics
            
            print(f"  Benchmark R²: Historical Mean: {benchmark_metrics['historical_mean_r2']:.4f}, "
                  f"Random Walk: {benchmark_metrics['random_walk_r2']:.4f}, "
                  f"Moving Avg: {benchmark_metrics['moving_avg_r2']:.4f}")
            
            # Initialize results for this feature set
            self.results[ticker][feature_set_name] = {}
            self.best_params[ticker][feature_set_name] = {}
            
            # Loop through each model
            for model_name, model in self.models.items():
                # Skip LSTM models for now, we'll handle them separately
                if model_name in ["LSTM"]:
                    continue
                
                try:
                    # Apply hyperparameter optimization for applicable models
                    if model_name == "KNN":
                        # Optimize KNN hyperparameters
                        knn_params = self.optimize_knn(X_train, y_train, ticker, feature_set_name)
                        model = KNeighborsRegressor(**knn_params)
                        # Store best parameters
                        self.best_params[ticker][feature_set_name][model_name] = knn_params
                    
                    elif model_name == "RandomForest":
                        # Optimize Random Forest hyperparameters
                        best_params = self.optimize_random_forest(X_train, y_train, ticker, feature_set_name)
                        model = RandomForestRegressor(**best_params)
                        # Store best parameters
                        self.best_params[ticker][feature_set_name][model_name] = best_params
                    
                    # Train model
                    model.fit(X_train, y_train)
                    
                    # Make predictions
                    y_pred_train = model.predict(X_train)
                    y_pred_test = model.predict(X_test)
                    
                    # Calculate metrics
                    metrics = self.calculate_metrics(y_train, y_pred_train, y_test, y_pred_test)
                    
                    # Store results
                    self.results[ticker][feature_set_name][model_name] = {
                        **metrics,
                        'model': model,
                        'feature_names': feature_names,  # Store the feature names used for this model
                        'y_pred_test': y_pred_test,  # Store predictions for later visualization
                        'test_dates': test_data['Date'].values,  # Store dates for plotting
                        'test_data': test_data,  # Store the test data for proper alignment in visualization
                        'benchmark_comparison': {  # Compare with benchmarks
                            'vs_hist_mean': metrics['test_r2'] - benchmark_metrics['historical_mean_r2'],
                            'vs_random_walk': metrics['test_r2'] - benchmark_metrics['random_walk_r2'],
                            'vs_moving_avg': metrics['test_r2'] - benchmark_metrics['moving_avg_r2']
                        }
                    }
                    
                    # Get feature importances for Random Forest
                    if model_name == 'RandomForest':
                        importances = model.feature_importances_
                        indices = np.argsort(importances)[::-1]
                        top_features = [(feature_names[i], importances[i]) for i in indices[:10]]
                        self.results[ticker][feature_set_name][model_name]['feature_importances'] = top_features
                    
                    print(f"    {model_name} - Test R²: {metrics['test_r2']:.4f} "
                          f"(vs. Hist Mean: {self.results[ticker][feature_set_name][model_name]['benchmark_comparison']['vs_hist_mean']:.4f}), "
                          f"Dir Accuracy: {metrics['test_dir_accuracy']:.4f}")
                
                except Exception as e:
                    print(f"    Error with {model_name}: {str(e)}")
                    # Instead of placeholders, just print the error and continue
                    continue
            
            # Handle LSTM model separately
            if "LSTM" in self.models and X_train.shape[0] > self.lookback:
                try:
                    # Prepare data for LSTM
                    X_train_lstm, y_train_lstm = self.prepare_lstm_data(X_train, y_train, self.lookback)
                    X_test_lstm, y_test_lstm = self.prepare_lstm_data(X_test, y_test, self.lookback)
                    
                    # Build and train LSTM model with ticker-specific hyperparameters
                    lstm_model = self.build_lstm_model(
                        (self.lookback, X_train.shape[1]), 
                        ticker=ticker, 
                        feature_set=feature_set_name
                    )
                    
                    # Only proceed if model was successfully created
                    if lstm_model is not None:
                        # Define callbacks
                        early_stopping = EarlyStopping(
                            monitor='val_loss',
                            patience=10,
                            restore_best_weights=True
                        )
                        
                        reduce_lr = ReduceLROnPlateau(
                            monitor='val_loss',
                            factor=0.2,
                            patience=5,
                            min_lr=1e-6
                        )
                        
                        # Check if we're using the fallback model
                        is_fallback = getattr(lstm_model, 'is_fallback', False)
                        
                        # Reshape input if using fallback model
                        if is_fallback:
                            X_train_lstm_reshaped = X_train_lstm.reshape(X_train_lstm.shape[0], -1)
                            X_test_lstm_reshaped = X_test_lstm.reshape(X_test_lstm.shape[0], -1)
                            
                            # Train the fallback model
                            history = lstm_model.fit(
                                X_train_lstm_reshaped, y_train_lstm,
                                epochs=100,
                                batch_size=32,
                                validation_split=0.2,
                                callbacks=[early_stopping, reduce_lr],
                                verbose=0
                            )
                            
                            # Make predictions
                            y_pred_train_lstm = lstm_model.predict(X_train_lstm_reshaped).flatten()
                            y_pred_test_lstm = lstm_model.predict(X_test_lstm_reshaped).flatten()
                        else:
                            # Train the regular LSTM model
                            history = lstm_model.fit(
                                X_train_lstm, y_train_lstm,
                                epochs=100,
                                batch_size=32,
                                validation_split=0.2,
                                callbacks=[early_stopping, reduce_lr],
                                verbose=0
                            )
                            
                            # Make predictions
                            y_pred_train_lstm = lstm_model.predict(X_train_lstm).flatten()
                            y_pred_test_lstm = lstm_model.predict(X_test_lstm).flatten()
                        
                        # Calculate metrics
                        metrics = self.calculate_metrics(y_train_lstm, y_pred_train_lstm, y_test_lstm, y_pred_test_lstm)
                        
                        # Adjust test_data to match LSTM predictions (remove first 'lookback' rows)
                        lstm_test_dates = test_data.iloc[self.lookback:].reset_index(drop=True)['Date'].values
                        lstm_test_data = test_data.iloc[self.lookback:].reset_index(drop=True)
                        
                        # Store results
                        self.results[ticker][feature_set_name]['LSTM'] = {
                            **metrics,
                            'model': lstm_model,
                            'feature_names': feature_names,
                            'history': history.history,
                            'is_fallback': is_fallback,
                            'y_pred_test': y_pred_test_lstm,
                            'test_dates': lstm_test_dates,
                            'test_data': lstm_test_data,  # Store the test data for proper alignment
                            'benchmark_comparison': {  # Compare with benchmarks
                                'vs_hist_mean': metrics['test_r2'] - benchmark_metrics['historical_mean_r2'],
                                'vs_random_walk': metrics['test_r2'] - benchmark_metrics['random_walk_r2'],
                                'vs_moving_avg': metrics['test_r2'] - benchmark_metrics['moving_avg_r2']
                            }
                        }
                        
                        # Store fixed hyperparameters
                        if is_fallback:
                            self.best_params[ticker][feature_set_name]['LSTM'] = {
                                'model_type': 'fallback_dense',
                                'units': 64,
                                'dropout': 0.2,
                                'learning_rate': 0.001
                            }
                        else:
                            self.best_params[ticker][feature_set_name]['LSTM'] = {
                                'units': 64,
                                'dropout': 0.2,
                                'activation': 'tanh',
                                'learning_rate': 0.001
                            }
                        
                        print(f"    LSTM - Test R²: {metrics['test_r2']:.4f} "
                              f"(vs. Hist Mean: {self.results[ticker][feature_set_name]['LSTM']['benchmark_comparison']['vs_hist_mean']:.4f}), "
                              f"Dir Accuracy: {metrics['test_dir_accuracy']:.4f}")
                    else:
                        print(f"    LSTM - Skipped due to model creation failure")
                except Exception as e:
                    print(f"    Error with LSTM: {str(e)}")
        
        # Save results and models for this ticker
        self.save_results(ticker)
    
    def plot_actual_vs_predicted(self, ticker):
        """
        Generate plots comparing actual vs. predicted returns for each model and feature set.
        
        Parameters:
        -----------
        ticker : str
            Stock ticker symbol
        """
        # Create directory for this ticker's plots
        ticker_dir = os.path.join(self.results_dir, ticker, 'plots')
        if not os.path.exists(ticker_dir):
            os.makedirs(ticker_dir)
        
        # Only use these feature sets
        desired_feature_sets = ["all_features", "price_only", "price_and_financial", "price_and_sentiment"]
        
        plot_count = 0
        
        # Generate plots for each feature set and model
        for feature_set, models in self.results[ticker].items():
            # Skip feature sets that are not in our desired list
            if feature_set not in desired_feature_sets:
                continue
                
            for model_name, results in models.items():
                # Skip models that failed
                if 'model' not in results or results['model'] is None or 'y_pred_test' not in results:
                    continue
                
                # Get predictions and actual values
                y_pred = results['y_pred_test']
                dates = results['test_dates']
                
                # Use the test_data stored with the results rather than rebuilding it
                if 'test_data' in results:
                    test_data = results['test_data']
                    y_actual = test_data['next_day_return'].values
                else:
                    # Fallback if test_data wasn't stored (although it should be)
                    print(f"Warning: Using fallback test data for {ticker}, {feature_set}, {model_name}")
                    test_data = self.ticker_data[ticker]
                    train_size = int(len(test_data) * 0.95)  # Using 95% train
                    test_data = test_data.iloc[train_size:].reset_index(drop=True)
                    
                    # Apply lookback offset for LSTM
                    if model_name == 'LSTM':
                        test_data = test_data.iloc[self.lookback:].reset_index(drop=True)
                    y_actual = test_data['next_day_return'].values
                
                # Make sure predictions and actuals have the same length
                if len(y_pred) != len(y_actual):
                    min_length = min(len(y_pred), len(y_actual))
                    y_pred = y_pred[:min_length]
                    y_actual = y_actual[:min_length]
                    dates = dates[:min_length]
                
                # Create the plot
                plt.figure(figsize=(12, 6))
                
                # Plot actual and predicted returns
                plt.plot(dates, y_actual, label='Actual Returns', color='blue')
                plt.plot(dates, y_pred, label='Predicted Returns', color='red')
                
                # Add title and labels
                plt.title(f'{ticker} - {model_name} with {feature_set} - Actual vs. Predicted Returns')
                plt.xlabel('Date')
                plt.ylabel('Return')
                plt.legend()
                plt.grid(True)
                
                # Format date axis
                plt.gcf().autofmt_xdate()
                
                # Add metrics to the plot
                benchmark_text = ''
                if 'benchmark_comparison' in results:
                    vs_hist_mean = results['benchmark_comparison']['vs_hist_mean']
                    vs_random_walk = results['benchmark_comparison']['vs_random_walk']
                    benchmark_text = f", R² vs. Hist Mean: {vs_hist_mean:.4f}, vs. Random Walk: {vs_random_walk:.4f}"
                
                plt.figtext(0.15, 0.02, 
                            f"R² = {results['test_r2']:.4f}, "
                            f"Dir Accuracy = {results.get('test_dir_accuracy', 0):.4f}"
                            f"{benchmark_text}",
                            ha='left', fontsize=10, 
                            bbox=dict(facecolor='white', alpha=0.8))
                
                # Save the plot
                plt.tight_layout()
                plt.savefig(os.path.join(ticker_dir, f'{ticker}_{feature_set}_{model_name}_predictions.png'))
                plt.close()
                
                plot_count += 1
        
        if plot_count > 0:
            print(f"Generated {plot_count} prediction plots for {ticker}")
        else:
            print(f"Warning: No plots were generated for {ticker}. Check if data is available.")
    
    def save_results(self, ticker):
        """
        Save results and models for a given ticker.
        
        Parameters:
        -----------
        ticker : str
            Stock ticker symbol
        """
        # Create directory for this ticker
        ticker_dir = os.path.join(self.results_dir, ticker)
        if not os.path.exists(ticker_dir):
            os.makedirs(ticker_dir)
        
        # Save results as CSV
        results_df = self.format_results_as_dataframe(ticker)
        results_df.to_csv(os.path.join(ticker_dir, f"{ticker}_results.csv"), index=False)
        
        # Save hyperparameters
        hyperparams_df = self.format_hyperparams_as_dataframe(ticker)
        hyperparams_df.to_csv(os.path.join(ticker_dir, f"{ticker}_hyperparameters.csv"), index=False)
        
        # Save benchmark metrics
        benchmark_df = self.format_benchmark_metrics_as_dataframe(ticker)
        benchmark_df.to_csv(os.path.join(ticker_dir, f"{ticker}_benchmark_metrics.csv"), index=False)
        
        # Save best models
        best_models = self.find_best_models(ticker)
        for feature_set, model_info in best_models.items():
            model_name = model_info['model_name']
            
            # Skip if model results don't exist
            if model_name not in self.results[ticker][feature_set]:
                continue
                
            model = self.results[ticker][feature_set][model_name]['model']
            
            # Skip saving models that failed
            if model is None:
                continue
                
            # Save feature names used for this model
            feature_names = self.results[ticker][feature_set][model_name].get('feature_names', [])
            feature_names_path = os.path.join(ticker_dir, f"{ticker}_{feature_set}_{model_name}_features.joblib")
            joblib.dump(feature_names, feature_names_path)
            
            # Save model
            if model_name in ["LSTM"]:
                # Save Keras model
                try:
                    model_path = os.path.join(ticker_dir, f"{ticker}_{feature_set}_{model_name}")
                    model.save(model_path)
                except Exception as e:
                    print(f"Error saving {model_name} model for {ticker}: {str(e)}")
            else:
                # Save scikit-learn model
                try:
                    model_path = os.path.join(ticker_dir, f"{ticker}_{feature_set}_{model_name}.joblib")
                    joblib.dump(model, model_path)
                except Exception as e:
                    print(f"Error saving {model_name} model for {ticker}: {str(e)}")
            
        print(f"Saved results and models for {ticker}")
    
    def format_results_as_dataframe(self, ticker):
        """
        Format results as a pandas DataFrame for easier analysis, including
        train/test directional accuracy.
        """
        rows = []
        for feature_set, models in self.results[ticker].items():
            for model_name, metrics in models.items():
                # Skip models without valid R²
                if 'test_r2' not in metrics or pd.isna(metrics['test_r2']):
                    continue

                # Copy all scalar metrics except the large objects
                skip = {
                    'model', 'feature_importances', 'feature_names', 'history',
                    'y_pred_test', 'test_dates', 'is_fallback',
                    'benchmark_comparison', 'test_data'
                }
                model_dict = {k: v for k, v in metrics.items() if k not in skip}

                # Add identifiers
                model_dict.update({
                    'ticker': ticker,
                    'feature_set': feature_set,
                    'model': model_name
                })
                
                # Add benchmark comparison values
                if 'benchmark_comparison' in metrics:
                    for key, value in metrics['benchmark_comparison'].items():
                        model_dict[key] = value
                
                rows.append(model_dict)

        # Ensure the columns exist even if there are no valid rows
        if not rows:
            columns = [
                'ticker', 'feature_set', 'model',
                'train_r2', 'test_r2', 
                'train_dir_accuracy', 'test_dir_accuracy',
                'vs_hist_mean', 'vs_random_walk', 'vs_moving_avg'
            ]
            return pd.DataFrame(columns=columns)

        return pd.DataFrame(rows)
    
    def format_hyperparams_as_dataframe(self, ticker):
        """
        Format hyperparameters as a pandas DataFrame.
        
        Parameters:
        -----------
        ticker : str
            Stock ticker symbol
            
        Returns:
        --------
        pandas.DataFrame
            Formatted hyperparameters
        """
        rows = []
        
        for feature_set, models in self.best_params[ticker].items():
            for model_name, params in models.items():
                # Convert param dict to string
                params_str = str(params)
                
                rows.append({
                    'ticker': ticker,
                    'feature_set': feature_set,
                    'model': model_name,
                    'hyperparameters': params_str
                })
        
        # If no hyperparameters, return empty DataFrame with expected columns
        if not rows:
            return pd.DataFrame(columns=['ticker', 'feature_set', 'model', 'hyperparameters'])
        
        return pd.DataFrame(rows)
    
    def format_benchmark_metrics_as_dataframe(self, ticker):
        """
        Format benchmark metrics as a pandas DataFrame.
        
        Parameters:
        -----------
        ticker : str
            Stock ticker symbol
            
        Returns:
        --------
        pandas.DataFrame
            Formatted benchmark metrics
        """
        rows = []
        
        for feature_set, metrics in self.benchmark_metrics[ticker].items():
            metrics_dict = metrics.copy()
            metrics_dict.update({
                'ticker': ticker,
                'feature_set': feature_set
            })
            rows.append(metrics_dict)
        
        # If no benchmark metrics, return empty DataFrame with expected columns
        if not rows:
            return pd.DataFrame(columns=['ticker', 'feature_set', 'historical_mean_r2', 
                                         'random_walk_r2', 'moving_avg_r2'])
        
        return pd.DataFrame(rows)
    
    def find_best_models(self, ticker):
        """
        Find the best model for each feature set based primarily on out-of-sample R².
        
        Parameters:
        -----------
        ticker : str
            Stock ticker symbol
            
        Returns:
        --------
        dict
            Best model for each feature set
        """
        best_models = {}
        
        for feature_set, models in self.results[ticker].items():
            best_r2 = float('-inf')
            best_model = None
            
            for model_name, metrics in models.items():
                # Skip models with missing metrics
                if 'test_r2' not in metrics or pd.isna(metrics['test_r2']):
                    continue
                
                # Use R² as the primary measure for model selection
                r2_score_value = metrics['test_r2']
                
                if r2_score_value > best_r2:
                    best_r2 = r2_score_value
                    best_model = model_name
            
            if best_model:
                best_models[feature_set] = {
                    'model_name': best_model,
                    'test_r2': models[best_model]['test_r2'],
                    'test_dir_accuracy': models[best_model]['test_dir_accuracy'],
                }
        
        return best_models
    
    def visualize_results(self, metric='test_r2'):
        """
        Visualize results across all tickers and feature sets.
        
        Parameters:
        -----------
        metric : str
            Metric to visualize
        """
        # --- Heatmap of average metric by feature_set/model ---
        plt.figure(figsize=(15, 10))
        
        # Prepare data for visualization
        all_results = []
        for ticker, ticker_results in self.results.items():
            for feature_set, models in ticker_results.items():
                for model_name, metrics in models.items():
                    if metric in metrics and not pd.isna(metrics[metric]):
                        all_results.append({
                            'ticker': ticker,
                            'feature_set': feature_set,
                            'model': model_name,
                            metric: metrics[metric]
                        })
        
        # Check if we have any results
        if not all_results:
            print(f"No valid results for metric '{metric}'. Skipping visualization.")
            return
            
        # Convert to DataFrame
        df = pd.DataFrame(all_results)
        
        # Create pivot table
        pivot = df.pivot_table(
            values=metric,
            index='feature_set',
            columns='model',
            aggfunc='mean'
        )
        
        # Plot heatmap
        sns.heatmap(pivot, annot=True, cmap='viridis', fmt='.4f')
        # title tweak for directional accuracy
        if metric == 'test_dir_accuracy':
            plt.title('Average Directional Accuracy across all tickers')
        else:
            plt.title(f'Average {metric} across all tickers')
        plt.tight_layout()
        plt.savefig(os.path.join(self.results_dir, f'overall_{metric}_heatmap.png'))
        
        # --- Boxplot of metric by feature_set/model ---
        plt.figure(figsize=(15, 10))
        sns.boxplot(x='feature_set', y=metric, hue='model', data=df)
        # title tweak for directional accuracy
        if metric == 'test_dir_accuracy':
            plt.title('Directional Accuracy by Feature Set and Model')
        else:
            plt.title(f'{metric} by Feature Set and Model')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.savefig(os.path.join(self.results_dir, f'overall_{metric}_boxplot.png'))
        
        # --- R²‐only benchmark comparisons (unchanged) ---
        if metric == 'test_r2':
            benchmark_columns = ['vs_hist_mean', 'vs_random_walk', 'vs_moving_avg']
            all_benchmark_results = []
            for ticker, ticker_results in self.results.items():
                for feature_set, models in ticker_results.items():
                    for model_name, metrics in models.items():
                        if 'benchmark_comparison' in metrics:
                            for benchmark_name, value in metrics['benchmark_comparison'].items():
                                if not pd.isna(value):
                                    all_benchmark_results.append({
                                        'ticker': ticker,
                                        'feature_set': feature_set,
                                        'model': model_name,
                                        'benchmark': benchmark_name,
                                        'improvement': value
                                    })
            
            if all_benchmark_results:
                benchmark_df = pd.DataFrame(all_benchmark_results)
                
                plt.figure(figsize=(15, 10))
                sns.boxplot(x='model', y='improvement', hue='benchmark', data=benchmark_df)
                plt.title('R² Improvement over Benchmarks by Model')
                plt.xticks(rotation=45)
                plt.tight_layout()
                plt.savefig(os.path.join(self.results_dir, 'benchmark_improvement_by_model.png'))
                
                plt.figure(figsize=(15, 10))
                sns.boxplot(x='feature_set', y='improvement', hue='benchmark', data=benchmark_df)
                plt.title('R² Improvement over Benchmarks by Feature Set')
                plt.xticks(rotation=45)
                plt.tight_layout()
                plt.savefig(os.path.join(self.results_dir, 'benchmark_improvement_by_feature_set.png'))
        
        print(f"Visualizations saved to {self.results_dir}")
    
    def save_results_to_file(self, filename='model_results.pkl'):
        """
        Save model results to a file to avoid rerunning the entire program.
        
        Parameters:
        -----------
        filename : str
            Name of the file to save results to
        """
        save_path = os.path.join(self.results_dir, filename)
        
        # Create a dictionary with all the data we want to save
        data_to_save = {
            'results': self.results,
            'best_params': self.best_params,
            'benchmark_metrics': self.benchmark_metrics,
            'feature_sets': self.feature_sets,
        }
        
        # Save using pickle
        with open(save_path, 'wb') as f:
            pickle.dump(data_to_save, f)
        
        print(f"Results saved to {save_path}")
    
    def load_results_from_file(self, filename='model_results.pkl'):
        """
        Load model results from a file.
        
        Parameters:
        -----------
        filename : str
            Name of the file to load results from
            
        Returns:
        --------
        bool
            True if loading was successful, False otherwise
        """
        load_path = os.path.join(self.results_dir, filename)
        
        if not os.path.exists(load_path):
            print(f"File {load_path} does not exist. Cannot load results.")
            return False
        
        try:
            # Load using pickle
            with open(load_path, 'rb') as f:
                data = pickle.load(f)
            
            # Restore data
            self.results = data['results']
            self.best_params = data['best_params']
            self.benchmark_metrics = data['benchmark_metrics']
            self.feature_sets = data.get('feature_sets', {})
            
            print(f"Results loaded from {load_path}")
            return True
        except Exception as e:
            print(f"Error loading results: {str(e)}")
            return False
    
    def run(self, save_results=True):
        """
        Run the complete stock selection model pipeline.
        
        Parameters:
        -----------
        save_results : bool
            Whether to save results to a file
        """
        try:
            start_time = time.time()
            
            # Load data
            self.load_data()
            
            if not self.ticker_data:
                print("No ticker data was loaded. Please check the data directory.")
                return
            
            # Create feature sets
            self.create_feature_sets()
            
            # Build models
            self.build_models()

            # Train and evaluate for each ticker
            for ticker in tqdm(list(self.ticker_data.keys()), desc="Processing tickers"):
                try:
                    self.train_and_evaluate(ticker)
                    # Generate plots for this ticker after evaluation
                    self.plot_actual_vs_predicted(ticker)
                except Exception as e:
                    print(f"Error evaluating ticker {ticker}: {str(e)}")
                    print(f"Skipping to next ticker...")
            
            # Check if we have any results
            if not self.results:
                print("No results were generated. Cannot create visualizations.")
                return
            
            try:
                # Visualize results for R² and directional accuracy
                self.visualize_results(metric='test_r2')
                self.visualize_results(metric='test_dir_accuracy')
            except Exception as e:
                print(f"Error creating visualizations: {str(e)}")
            
            # Save results if requested
            if save_results:
                self.save_results_to_file()
            
            end_time = time.time()
            execution_time = end_time - start_time
            print(f"\nStock selection model analysis completed in {execution_time:.2f} seconds!")
        
        except Exception as e:
            print(f"Error in main execution: {str(e)}")
            import traceback
            traceback.print_exc()


class PortfolioSimulator:
    def __init__(self, results, tickers,
                 num_positions=30,
                 neutral_band=0.0):
        """
        results: model.results
        tickers: list(model.results.keys())
        num_positions: how many names to pick each day (long+short)
        neutral_band: abs(pred) below this ⇒ skip that ticker
        """
        self.results   = results
        self.tickers   = tickers
        self.N         = num_positions
        self.neutral   = neutral_band

    def simulate(self, feature_set, model_name, weight_by_sign=True):
        """
        weight_by_sign=False ⇒ old 50/50 long/short
        weight_by_sign=True  ⇒ equal $ per ticker, signed by direction
        """
        import pandas as pd
        dfs = []
        for t in self.tickers:
            # Check if the ticker has the requested feature set and model
            if feature_set not in self.results[t] or model_name not in self.results[t][feature_set]:
                print(f"Warning: {t} does not have data for {feature_set}|{model_name}. Skipping.")
                continue
                
            info = self.results[t][feature_set][model_name]
            
            # Check if required data is present
            if 'y_pred_test' not in info or 'test_dates' not in info or 'test_data' not in info:
                print(f"Warning: Missing required data for {t} with {feature_set}|{model_name}. Skipping.")
                continue
                
            # Debug print to check if data exists for this combination
            print(f"Processing {t} with {feature_set}|{model_name}. Data points: {len(info['y_pred_test'])}")
            
            df = pd.DataFrame({
                'pred': info['y_pred_test'],
                'ret' : info['test_data']['next_day_return'].values
            }, index=pd.to_datetime(info['test_dates']))
            df['ticker'] = t
            dfs.append(df)
            
        if not dfs:
            print(f"Error: No valid data for {feature_set}|{model_name}. Cannot simulate portfolio.")
            return pd.DataFrame({'daily': [], 'cum': []})
            
        all_df = pd.concat(dfs)

        port = []
        for date, grp in all_df.groupby(level=0):
            # apply neutral filter
            grp = grp.loc[grp['pred'].abs() >= self.neutral]
            if len(grp) < self.N:
                port.append((date, 0.0))
                continue

            # pick the top-N by confidence
            grp = grp.reindex(grp['pred'].abs().sort_values(ascending=False).index)[:self.N]

            if not weight_by_sign:
                # classic 50/50: half capital into longs, half into shorts
                half = self.N // 2
                longs  = grp.nlargest(half, 'pred')['ret']
                shorts = grp.nsmallest(half, 'pred')['ret']
                daily_ret = 0.5 * (longs.mean() - shorts.mean())
            else:
                # equal $ per pick, long if pred>0, short if pred<0
                signs   = grp['pred'].apply(lambda x: 1 if x>0 else -1)
                weights = signs / signs.abs().sum()
                daily_ret = (weights * grp['ret']).sum()

            port.append((date, daily_ret))

        s = pd.Series({d: r for d, r in port}).sort_index()
        cum = (1 + s).cumprod() - 1.0
        return pd.DataFrame({'daily': s, 'cum': cum})

    def simulate_all(self, weight_by_sign=False):
        """
        Simulate portfolio for all feature set and model combinations.
        
        Parameters:
        -----------
        weight_by_sign : bool
            Whether to weight by sign
            
        Returns:
        --------
        pandas.DataFrame
            DataFrame with cumulative returns for each feature set and model combination
        """
        combos = []
        
        # Get all available combinations
        for t in self.tickers:
            for fs in self.results[t]:
                for m in self.results[t][fs]:
                    combo = (fs, m)
                    if combo not in combos:
                        combos.append(combo)
        
        print(f"Found {len(combos)} feature set and model combinations")
        
        out = {}
        for fs, m in combos:
            print(f"Simulating portfolio for {fs}|{m}")
            df = self.simulate(fs, m, weight_by_sign=weight_by_sign)
            if not df.empty:
                out[f"{fs}|{m}"] = df['cum']
            else:
                print(f"No results for {fs}|{m}. Skipping.")
        
        return pd.DataFrame(out)
    
    def save_results_to_csv(self, df_cum, filename='portfolio_results.csv'):
        """
        Save portfolio simulation results to a CSV file.
        
        Parameters:
        -----------
        df_cum : pandas.DataFrame
            DataFrame with cumulative returns
        filename : str
            Name of the file to save results to
        """
        df_cum.to_csv(filename)
        print(f"Portfolio results saved to {filename}")


if __name__ == "__main__":
    # Check if results file exists and load if possible
    model = StockSelectionModel(
        data_dir="processed_data/", 
        results_dir="model_results_v11/",
        use_lstm=True,  # Enable LSTM
        risk_free_rate=0.045,  # 4.5% annual risk-free rate
        trading_days_per_year=252  # Standard trading days per year
    )
    
    # Try to load results from file
    results_loaded = model.load_results_from_file()
    
    # If results weren't loaded, run the model
    if not results_loaded:
        print("Results not found. Running the model...")
        model.run(save_results=True)
    else:
        print("Results loaded successfully. Skipping model training.")
    
    # Create and run portfolio simulator
    sim = PortfolioSimulator(
        results=model.results,
        tickers=list(model.results.keys()),
        num_positions=30,        # total names/day
        neutral_band=0.0         # skip preds near zero
    )
    
    # Get a DataFrame of cumulative returns for every (feature_set,model)
    df_cum = sim.simulate_all()
    
    # Save portfolio results to CSV
    sim.save_results_to_csv(df_cum, os.path.join(model.results_dir, "portfolio_cumulative_returns.csv"))
    
    # 1) Pick only the columns we care about
    desired_fs = [
        "price_only",
        "price_and_sentiment",
        "price_and_financial",
        "all_features"
    ]
    cols = [c for c in df_cum.columns if c.split("|")[0] in desired_fs]
    df_cum_fs = df_cum[cols]
    
    # 2) One plot with ALL feature-sets together (averaged across models)
    avg_df = pd.DataFrame({
        fs: df_cum_fs.filter(like=f"{fs}|").mean(axis=1)
        for fs in desired_fs
    })
    plt.figure(figsize=(12,6))
    for fs in desired_fs:
        if fs in avg_df.columns:
            plt.plot(avg_df.index, avg_df[fs], label=fs)
    plt.title("Average Cumulative Returns by Feature Set")
    plt.xlabel("Date")
    plt.ylabel("Cumulative Return")
    plt.legend()
    plt.tight_layout()
    plt.savefig(os.path.join(model.results_dir, "cumulative_all_features_together.png"))
    plt.close()
    
    # 3) One plot per model, comparing the four feature-sets
    for m in sorted({c.split("|")[1] for c in df_cum_fs.columns}):
        available = [fs for fs in desired_fs if f"{fs}|{m}" in df_cum_fs.columns]
        if not available:
            print(f"No data available for model {m}. Skipping plot.")
            continue
        
        plt.figure(figsize=(12,6))
        for fs in available:
            col_name = f"{fs}|{m}"
            if col_name in df_cum_fs.columns:
                plt.plot(
                    df_cum_fs.index,
                    df_cum_fs[col_name],
                    label=fs
                )
            else:
                print(f"Warning: {col_name} not found in df_cum_fs.columns")
        
        plt.title(f"Cumulative Returns for {m} — Feature Set Comparison")
        plt.xlabel("Date")
        plt.ylabel("Cumulative Return")
        plt.legend()
        plt.tight_layout()
        plt.savefig(
            os.path.join(
                model.results_dir,
                f"cumulative_{m}_by_feature_set.png"
            )
        )
        plt.close()
    
    # 4) Table of final cumulative returns (last date) for each feature & model
    # Build a flat DataFrame from your dict of (model,feature_set) → value
    final_vals = {}
    for col in df_cum_fs.columns:
        if not df_cum_fs[col].empty:
            model_name = col.split("|")[1]
            feature_set = col.split("|")[0]
            final_vals[(model_name, feature_set)] = df_cum_fs[col].iloc[-1]
    
    if final_vals:
        table = pd.DataFrame.from_dict(
            final_vals,
            orient='index',
            columns=['final_cum_return']
        )
        
        # Convert that flat Index of 2-tuples into a true MultiIndex with two levels
        table.index = pd.MultiIndex.from_tuples(
            table.index,
            names=['model', 'feature_set']
        )
        
        # Pivot so rows=model, columns=feature_set
        final_returns = (
            table
            .reset_index()
            .pivot(index='model', columns='feature_set', values='final_cum_return')
        )
        
        # Save and print
        final_returns.to_csv(os.path.join(model.results_dir, "final_returns_table.csv"))
        print(final_returns)
    else:
        print("No final values available to create the final returns table.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
from matplotlib.colors import LinearSegmentedColormap

# Define the feature sets we want to include
FEATURE_SETS = ["price_only", "price_and_financial", "price_and_sentiment", "all_features"]

def load_ticker_data(ticker, base_path="model_results_v10"):
    """
    Load data for a single ticker
    """
    file_path = f"{base_path}/{ticker}/{ticker}_results.csv"
    try:
        df = pd.read_csv(file_path)
        # Filter to include only the requested feature sets
        df = df[df['feature_set'].isin(FEATURE_SETS)]
        return df
    except Exception as e:
        print(f"Error loading data for {ticker}: {str(e)}")
        return None

def process_all_ticker_data(base_path="model_results_v10"):
    """
    Process data from all ticker folders
    """
    # Get all ticker folders
    ticker_folders = glob.glob(f"{base_path}/*")
    tickers = sorted([os.path.basename(folder) for folder in ticker_folders])
    
    print(f"Found {len(tickers)} ticker folders")
    
    # If no tickers found, try using the AAPL_results1.csv directly
    if not tickers and os.path.exists("AAPL_results1.csv"):
        print("No ticker folders found, using AAPL_results1.csv file directly")
        df = pd.read_csv("AAPL_results1.csv")
        df = df[df['feature_set'].isin(FEATURE_SETS)]
        return df, ["AAPL"]
    
    # Load data for all tickers
    all_data = []
    valid_tickers = []
    
    for ticker in tickers:
        df = load_ticker_data(ticker, base_path)
        if df is not None and not df.empty:
            all_data.append(df)
            valid_tickers.append(ticker)
            print(f"Loaded data for {ticker}")
    
    if not all_data:
        raise ValueError("No data was loaded. Check the path and file structure.")
    
    # Combine all data
    combined_data = pd.concat(all_data, ignore_index=True)
    return combined_data, valid_tickers

def create_r2_heatmap_separate(data, tickers, output_path="r2_heatmap_large.png"):
    """
    Create a separate, larger R² heatmap with rotated text
    """
    # Create a combined model + feature set column
    data['model_feature'] = data['model'] + ' | ' + data['feature_set']
    
    # Sort models in a logical order
    model_order = ['OLS', 'Ridge', 'Lasso', 'KNN', 'RandomForest', 'LSTM']
    
    # Create combined model_feature order
    model_feature_order = []
    for model in model_order:
        for feature in FEATURE_SETS:
            model_feature_order.append(f"{model} | {feature}")
    
    # Create pivot table for R²
    r2_pivot = data.pivot_table(index='model_feature', columns='ticker', values='test_r2')
    
    # Filter to only include model_features that exist in the data
    valid_model_features = [mf for mf in model_feature_order if mf in r2_pivot.index]
    
    # Reindex to maintain desired order
    r2_pivot = r2_pivot.reindex(valid_model_features)
    
    # Create a copy of the data with capped values for better visualization
    r2_display = r2_pivot.copy()
    
    # Handle extreme values
    # Mark cells with extreme values
    extreme_mask = (np.abs(r2_pivot) > 5)
    
    # Cap values for display purposes
    r2_display[r2_display < -1] = -1
    r2_display[r2_display > 1] = 1
    
    # Create figure with appropriate size
    height = max(16, len(valid_model_features) * 0.5)
    width = max(14, len(tickers) * 0.7)
    
    plt.figure(figsize=(width, height))
    
    # Create heatmap with pre-capped values for visualization
    ax = sns.heatmap(
        r2_display,
        annot=False,  # No automatic annotations
        cmap='RdBu_r',
        center=0,
        vmin=-1,
        vmax=1,
        cbar_kws={'label': 'R² Value'}
    )
    
    # Add separators between model groups
    for i in range(4, len(valid_model_features), 4):
        ax.axhline(y=i, color='gray', linestyle='-', alpha=0.5)
    
    # Add a marker for extreme values
    for i in range(len(r2_pivot.index)):
        for j in range(len(r2_pivot.columns)):
            if extreme_mask.iloc[i, j]:
                ax.add_patch(plt.Rectangle((j, i), 1, 1, fill=False, edgecolor='black', lw=1, linestyle=':'))
    
    # Now add text annotations with 90 degree rotation
    for i in range(len(r2_pivot.index)):
        for j in range(len(r2_pivot.columns)):
            value = r2_pivot.iloc[i, j]
            if pd.isna(value):
                continue
            
            # Format text based on value magnitude
            if abs(value) > 100:
                text = f"{value:.1e}"
                fontsize = 7
            elif abs(value) > 10:
                text = f"{value:.1e}"
                fontsize = 7.5
            elif abs(value) > 1:
                text = f"{value:.2f}"
                fontsize = 8
            else:
                text = f"{value:.2f}"
                fontsize = 8.5
            
            # Set color based on background darkness
            display_value = r2_display.iloc[i, j]
            text_color = 'white' if abs(display_value) > 0.6 else 'black'
            
            # Add rotated text
            ax.text(
                j + 0.5,  # Center of cell (x)
                i + 0.5,  # Center of cell (y)
                text,
                ha='center',
                va='center',
                rotation=90,  # Vertical text
                color=text_color,
                fontsize=fontsize
            )
    
    plt.title('Test R² Values by Model/Feature Combination and Ticker', fontsize=18)
    plt.ylabel('Model | Feature Set', fontsize=14)
    plt.xlabel('Ticker', fontsize=14)
    
    # Set rotation for tick labels
    plt.xticks(rotation=45)
    for tick in ax.get_xticklabels():
        tick.set_horizontalalignment('right')
    
    # Adjust layout
    plt.tight_layout()
    
    # Save figure
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    print(f"Saved R² heatmap with rotated text to '{output_path}'")
    plt.close()

def create_dir_accuracy_heatmap_separate(data, tickers, output_path="dir_accuracy_heatmap_large.png"):
    """
    Create a separate, larger direction accuracy heatmap with rotated text
    """
    # Create a combined model + feature set column
    data['model_feature'] = data['model'] + ' | ' + data['feature_set']
    
    # Sort models in a logical order
    model_order = ['OLS', 'Ridge', 'Lasso', 'KNN', 'RandomForest', 'LSTM']
    
    # Create combined model_feature order
    model_feature_order = []
    for model in model_order:
        for feature in FEATURE_SETS:
            model_feature_order.append(f"{model} | {feature}")
    
    # Create pivot table for direction accuracy
    dir_acc_pivot = data.pivot_table(index='model_feature', columns='ticker', values='test_dir_accuracy')
    
    # Filter to only include model_features that exist in the data
    valid_model_features = [mf for mf in model_feature_order if mf in dir_acc_pivot.index]
    
    # Reindex to maintain desired order
    dir_acc_pivot = dir_acc_pivot.reindex(valid_model_features)
    
    # Create figure with appropriate size
    height = max(16, len(valid_model_features) * 0.5)
    width = max(14, len(tickers) * 0.7)
    
    plt.figure(figsize=(width, height))
    
    # Create heatmap without annotations
    ax = sns.heatmap(
        dir_acc_pivot,
        annot=False,  # No automatic annotations
        cmap='RdYlGn',
        vmin=0.3,
        vmax=0.9,
        cbar_kws={'label': 'Direction Accuracy'}
    )
    
    # Add separators between model groups
    for i in range(4, len(valid_model_features), 4):
        ax.axhline(y=i, color='gray', linestyle='-', alpha=0.5)
    
    # Add rotated text annotations
    for i in range(len(dir_acc_pivot.index)):
        for j in range(len(dir_acc_pivot.columns)):
            value = dir_acc_pivot.iloc[i, j]
            if pd.isna(value):
                continue
            
            # Format as percentage
            text = f"{value:.1%}"
            
            # Set color based on value (for readability against background)
            text_color = 'black'
            if value < 0.4 or value > 0.75:
                text_color = 'white'
            
            # Add rotated text
            ax.text(
                j + 0.5,  # Center of cell (x)
                i + 0.5,  # Center of cell (y)
                text,
                ha='center',
                va='center',
                rotation=90,  # Vertical text
                color=text_color,
                fontsize=9
            )
    
    plt.title('Test Direction Accuracy by Model/Feature Combination and Ticker', fontsize=18)
    plt.ylabel('Model | Feature Set', fontsize=14)
    plt.xlabel('Ticker', fontsize=14)
    
    # Set rotation for tick labels
    plt.xticks(rotation=45)
    for tick in ax.get_xticklabels():
        tick.set_horizontalalignment('right')
    
    # Adjust layout
    plt.tight_layout()
    
    # Save figure
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    print(f"Saved direction accuracy heatmap with rotated text to '{output_path}'")
    plt.close()

def create_model_specific_heatmaps(data, tickers, output_dir="model_heatmaps"):
    """
    Create separate heatmaps for each model type with rotated text
    """
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    # Get unique models
    models = sorted(data['model'].unique())
    
    for model in models:
        # Filter data for this model
        model_data = data[data['model'] == model]
        
        # R² heatmap
        r2_pivot = model_data.pivot_table(index='feature_set', columns='ticker', values='test_r2')
        
        # Direction accuracy heatmap
        dir_acc_pivot = model_data.pivot_table(index='feature_set', columns='ticker', values='test_dir_accuracy')
        
        # Sort feature sets
        r2_pivot = r2_pivot.reindex(FEATURE_SETS)
        dir_acc_pivot = dir_acc_pivot.reindex(FEATURE_SETS)
        
        # Create figure
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(max(10, len(tickers) * 0.7), 10))
        
        # Set figure title
        fig.suptitle(f'{model} Model Performance', fontsize=18)
        
        # For R² heatmap
        # Handle LSTM separately due to extreme values
        if model == 'LSTM':
            # Create capped version for display
            r2_display = r2_pivot.copy()
            r2_display[r2_display < -5] = -5
            r2_display[r2_display > 5] = 5
            
            sns.heatmap(
                r2_display,
                ax=ax1,
                annot=False,  # Will add rotated text manually
                cmap='RdBu_r',
                center=0,
                vmin=-5,
                vmax=5,
                cbar_kws={'label': 'R² Value'}
            )
            
            # Add rotated text annotations
            for i in range(len(r2_pivot.index)):
                for j in range(len(r2_pivot.columns)):
                    value = r2_pivot.iloc[i, j]
                    if pd.isna(value):
                        continue
                    
                    # Format text based on value magnitude
                    if abs(value) > 100:
                        text = f"{value:.1e}"
                        fontsize = 7
                    elif abs(value) > 10:
                        text = f"{value:.1e}"
                        fontsize = 7.5
                    else:
                        text = f"{value:.2f}"
                        fontsize = 8
                    
                    # Set color based on background darkness
                    display_value = r2_display.iloc[i, j]
                    text_color = 'white' if abs(display_value) > 3 else 'black'
                    
                    # Add rotated text
                    ax1.text(
                        j + 0.5,  # Center of cell (x)
                        i + 0.5,  # Center of cell (y)
                        text,
                        ha='center',
                        va='center',
                        rotation=90,  # Vertical text
                        color=text_color,
                        fontsize=fontsize
                    )
        else:
            # For other models, use standard scaling
            sns.heatmap(
                r2_pivot,
                ax=ax1,
                annot=False,
                cmap='RdBu_r',
                center=0,
                vmin=-1,
                vmax=1,
                cbar_kws={'label': 'R² Value'}
            )
            
            # Add rotated text annotations
            for i in range(len(r2_pivot.index)):
                for j in range(len(r2_pivot.columns)):
                    value = r2_pivot.iloc[i, j]
                    if pd.isna(value):
                        continue
                    
                    # Format text
                    text = f"{value:.2f}"
                    
                    # Set color based on value
                    text_color = 'white' if abs(value) > 0.6 else 'black'
                    
                    # Add rotated text
                    ax1.text(
                        j + 0.5,
                        i + 0.5,
                        text,
                        ha='center',
                        va='center',
                        rotation=90,
                        color=text_color,
                        fontsize=9
                    )
        
        ax1.set_title(f'Test R² Values ({model})', fontsize=14)
        ax1.set_ylabel('Feature Set', fontsize=12)
        ax1.set_xlabel('')
        
        # Set rotation for tick labels
        ax1.tick_params(axis='x', rotation=45)
        for tick in ax1.get_xticklabels():
            tick.set_horizontalalignment('right')
        
        # For direction accuracy heatmap
        sns.heatmap(
            dir_acc_pivot,
            ax=ax2,
            annot=False,  # Will add rotated text manually
            cmap='RdYlGn',
            vmin=0.3,
            vmax=0.9,
            cbar_kws={'label': 'Direction Accuracy'}
        )
        
        # Add rotated text annotations
        for i in range(len(dir_acc_pivot.index)):
            for j in range(len(dir_acc_pivot.columns)):
                value = dir_acc_pivot.iloc[i, j]
                if pd.isna(value):
                    continue
                
                # Format as percentage
                text = f"{value:.1%}"
                
                # Set color based on value
                text_color = 'black'
                if value < 0.4 or value > 0.75:
                    text_color = 'white'
                
                # Add rotated text
                ax2.text(
                    j + 0.5,
                    i + 0.5,
                    text,
                    ha='center',
                    va='center',
                    rotation=90,
                    color=text_color,
                    fontsize=9
                )
        
        ax2.set_title(f'Test Direction Accuracy ({model})', fontsize=14)
        ax2.set_ylabel('Feature Set', fontsize=12)
        ax2.set_xlabel('Ticker', fontsize=12)
        
        # Set rotation for tick labels - fixed to avoid error
        ax2.tick_params(axis='x', rotation=45)
        for tick in ax2.get_xticklabels():
            tick.set_horizontalalignment('right')
        
        # Adjust layout
        plt.tight_layout()
        plt.subplots_adjust(top=0.9)
        
        # Save figure
        plt.savefig(f"{output_dir}/{model}_heatmap.png", dpi=300, bbox_inches='tight')
        plt.close()
        print(f"Created heatmap for {model} model with rotated text")

def create_summary_bar_charts(data, output_path="model_summary_charts.png"):
    """
    Create summary bar charts showing average performance by model/feature combination
    """
    # Calculate average R² and direction accuracy by model and feature set
    avg_r2 = data.groupby(['model', 'feature_set'])['test_r2'].mean().reset_index()
    avg_dir_acc = data.groupby(['model', 'feature_set'])['test_dir_accuracy'].mean().reset_index()
    
    # Create model_feature column
    avg_r2['model_feature'] = avg_r2['model'] + ' | ' + avg_r2['feature_set']
    avg_dir_acc['model_feature'] = avg_dir_acc['model'] + ' | ' + avg_dir_acc['feature_set']
    
    # Sort models in a logical order
    model_order = ['OLS', 'Ridge', 'Lasso', 'KNN', 'RandomForest', 'LSTM']
    
    # Create combined model_feature order
    model_feature_order = []
    for model in model_order:
        for feature in FEATURE_SETS:
            model_feature_order.append(f"{model} | {feature}")
    
    # Create figure
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 12))
    
    # Reorder rows
    valid_model_features = [mf for mf in model_feature_order if mf in avg_r2['model_feature'].values]
    
    # Filter and sort the DataFrames
    avg_r2_sorted = avg_r2[avg_r2['model_feature'].isin(valid_model_features)].copy()
    avg_r2_sorted['model_feature_order'] = avg_r2_sorted['model_feature'].map({mf: i for i, mf in enumerate(valid_model_features)})
    avg_r2_sorted = avg_r2_sorted.sort_values('model_feature_order')
    
    avg_dir_acc_sorted = avg_dir_acc[avg_dir_acc['model_feature'].isin(valid_model_features)].copy()
    avg_dir_acc_sorted['model_feature_order'] = avg_dir_acc_sorted['model_feature'].map({mf: i for i, mf in enumerate(valid_model_features)})
    avg_dir_acc_sorted = avg_dir_acc_sorted.sort_values('model_feature_order')
    
    # Cap extreme R² values for LSTM for better visualization
    avg_r2_sorted['test_r2_capped'] = avg_r2_sorted['test_r2'].copy()
    avg_r2_sorted.loc[avg_r2_sorted['test_r2'] < -5, 'test_r2_capped'] = -5
    
    # Create horizontal bar chart for R²
    bars1 = ax1.barh(
        avg_r2_sorted['model_feature'],
        avg_r2_sorted['test_r2_capped'],
        color=plt.cm.RdBu_r((avg_r2_sorted['test_r2_capped'] + 5) / 10)  # Normalize to 0-1 range for colormap
    )
    
    # Add text annotations
    for i, (idx, row) in enumerate(avg_r2_sorted.iterrows()):
        # Show actual value, not capped value
        value = row['test_r2']
        if abs(value) > 10:
            text = f"{value:.1e}"
        else:
            text = f"{value:.3f}"
        
        # Position the text
        if value < -5:
            # For extremely negative values, place text to the right of the bar
            ax1.text(-4.9, i, f"({text})", va='center', fontsize=9, color='black')
        else:
            # Normal placement
            text_x = max(-4.9, value - 0.5) if value < 0 else max(0.1, value + 0.1)
            ax1.text(text_x, i, text, va='center', ha='left' if value < 0 else 'right', fontsize=10)
    
    ax1.set_title('Average Test R² by Model & Feature Combination', fontsize=16)
    ax1.set_xlabel('R² Value (capped at -5 for extreme values)', fontsize=12)
    ax1.grid(True, linestyle='--', alpha=0.6)
    ax1.axvline(x=0, color='black', linestyle='-', alpha=0.7)
    ax1.set_xlim(-5.1, 1)
    
    # Add separators between model groups
    for i in range(4, len(valid_model_features), 4):
        ax1.axhline(y=i - 0.5, color='gray', linestyle='-', alpha=0.5, linewidth=1)
    
    # For direction accuracy
    bars2 = ax2.barh(
        avg_dir_acc_sorted['model_feature'],
        avg_dir_acc_sorted['test_dir_accuracy'],
        color=plt.cm.RdYlGn((avg_dir_acc_sorted['test_dir_accuracy'] - 0.3) / 0.6)  # Normalize to 0-1 range for colormap
    )
    
    # Add text annotations
    for i, (idx, row) in enumerate(avg_dir_acc_sorted.iterrows()):
        ax2.text(
            row['test_dir_accuracy'] + 0.01,
            i,
            f"{row['test_dir_accuracy']:.1%}",
            va='center',
            fontsize=10
        )
    
    ax2.set_title('Average Test Direction Accuracy by Model & Feature Combination', fontsize=16)
    ax2.set_xlabel('Direction Accuracy', fontsize=12)
    ax2.grid(True, linestyle='--', alpha=0.6)
    ax2.axvline(x=0.5, color='black', linestyle='--', alpha=0.7)  # Mark 50% (random guessing)
    ax2.set_xlim(0.3, 0.9)
    
    # Add separators between model groups
    for i in range(4, len(valid_model_features), 4):
        ax2.axhline(y=i - 0.5, color='gray', linestyle='-', alpha=0.5, linewidth=1)
    
    # Set y-axis labels only for the first plot to avoid redundancy
    ax2.set_yticklabels([])
    
    # Add legend for model groups
    patches = []
    for model in model_order:
        patches.append(plt.Rectangle((0, 0), 1, 1, fill=False, edgecolor='black', label=model))
    
    # Place legend at the bottom
    fig.legend(handles=patches, loc='lower center', ncol=len(model_order), bbox_to_anchor=(0.5, 0.02))
    
    # Adjust layout
    plt.tight_layout()
    plt.subplots_adjust(bottom=0.08)
    
    # Save figure
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    print(f"Saved summary bar charts to '{output_path}'")
    plt.close()

def main():
    """
    Main function to load data and create visualizations
    """
    try:
        # Specify the base path to the ticker folders
        base_path = "model_results_v10"
        
        # Process ticker data
        print("Processing ticker data...")
        data, tickers = process_all_ticker_data(base_path)
        
        print(f"Creating visualizations for {len(tickers)} tickers...")
        
        # Create separate large heatmaps with rotated text
        create_r2_heatmap_separate(data, tickers)
        create_dir_accuracy_heatmap_separate(data, tickers)
        
        # Create model-specific heatmaps with rotated text
        create_model_specific_heatmaps(data, tickers)
        
        # Create summary bar charts
        create_summary_bar_charts(data)
        
        print("Process completed successfully!")
        
    except Exception as e:
        print(f"Error: {str(e)}")
        import traceback
        traceback.print_exc()

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import os
import glob

# Define the feature sets we want to include
FEATURE_SETS = ["price_only", "price_and_financial", "price_and_sentiment", "all_features"]

def load_ticker_data(ticker, base_path="model_results_v10"):
    """
    Load data for a single ticker
    """
    file_path = f"{base_path}/{ticker}/{ticker}_results.csv"
    try:
        df = pd.read_csv(file_path)
        # Filter to include only the requested feature sets
        df = df[df['feature_set'].isin(FEATURE_SETS)]
        return df
    except Exception as e:
        print(f"Error loading data for {ticker}: {str(e)}")
        return None

def process_all_ticker_data(base_path="model_results_v10"):
    """
    Process data from all ticker folders
    """
    # Get all ticker folders
    ticker_folders = glob.glob(f"{base_path}/*")
    tickers = sorted([os.path.basename(folder) for folder in ticker_folders])
    
    print(f"Found {len(tickers)} ticker folders")
    
    # If no tickers found, try using the AAPL_results1.csv directly
    if not tickers and os.path.exists("AAPL_results1.csv"):
        print("No ticker folders found, using AAPL_results1.csv file directly")
        df = pd.read_csv("AAPL_results1.csv")
        df = df[df['feature_set'].isin(FEATURE_SETS)]
        return df, ["AAPL"]
    
    # Load data for all tickers
    all_data = []
    valid_tickers = []
    
    for ticker in tickers:
        df = load_ticker_data(ticker, base_path)
        if df is not None and not df.empty:
            all_data.append(df)
            valid_tickers.append(ticker)
            print(f"Loaded data for {ticker}")
    
    if not all_data:
        raise ValueError("No data was loaded. Check the path and file structure.")
    
    # Combine all data
    combined_data = pd.concat(all_data, ignore_index=True)
    return combined_data, valid_tickers

def export_direction_accuracy_csv(data, tickers, output_file="direction_accuracy_values.csv"):
    """
    Extract direction accuracy values and save to CSV
    """
    # Create a combined model + feature set column
    data['model_feature'] = data['model'] + ' | ' + data['feature_set']
    
    # Sort models in a logical order
    model_order = ['OLS', 'Ridge', 'Lasso', 'KNN', 'RandomForest', 'LSTM']
    
    # Create combined model_feature order
    model_feature_order = []
    for model in model_order:
        for feature in FEATURE_SETS:
            model_feature_order.append(f"{model} | {feature}")
    
    # Create pivot table for direction accuracy
    dir_acc_pivot = data.pivot_table(index='model_feature', columns='ticker', values='test_dir_accuracy')
    
    # Filter to only include model_features that exist in the data
    valid_model_features = [mf for mf in model_feature_order if mf in dir_acc_pivot.index]
    
    # Reindex to maintain desired order
    dir_acc_pivot = dir_acc_pivot.reindex(valid_model_features)
    
    # Add model and feature set as separate columns
    dir_acc_df = dir_acc_pivot.reset_index()
    dir_acc_df[['model', 'feature_set']] = dir_acc_df['model_feature'].str.split(' \| ', expand=True)
    
    # Reorder columns to put model and feature_set first
    cols = ['model', 'feature_set', 'model_feature'] + [col for col in dir_acc_df.columns if col not in ['model', 'feature_set', 'model_feature']]
    dir_acc_df = dir_acc_df[cols]
    
    # Save to CSV
    dir_acc_df.to_csv(output_file, index=False)
    print(f"Direction accuracy values saved to '{output_file}'")
    
    return dir_acc_df

def export_r2_values_csv(data, tickers, output_file="r2_values.csv"):
    """
    Extract R² values and save to CSV
    """
    # Create a combined model + feature set column
    data['model_feature'] = data['model'] + ' | ' + data['feature_set']
    
    # Sort models in a logical order
    model_order = ['OLS', 'Ridge', 'Lasso', 'KNN', 'RandomForest', 'LSTM']
    
    # Create combined model_feature order
    model_feature_order = []
    for model in model_order:
        for feature in FEATURE_SETS:
            model_feature_order.append(f"{model} | {feature}")
    
    # Create pivot table for R²
    r2_pivot = data.pivot_table(index='model_feature', columns='ticker', values='test_r2')
    
    # Filter to only include model_features that exist in the data
    valid_model_features = [mf for mf in model_feature_order if mf in r2_pivot.index]
    
    # Reindex to maintain desired order
    r2_pivot = r2_pivot.reindex(valid_model_features)
    
    # Add model and feature set as separate columns
    r2_df = r2_pivot.reset_index()
    r2_df[['model', 'feature_set']] = r2_df['model_feature'].str.split(' \| ', expand=True)
    
    # Reorder columns to put model and feature_set first
    cols = ['model', 'feature_set', 'model_feature'] + [col for col in r2_df.columns if col not in ['model', 'feature_set', 'model_feature']]
    r2_df = r2_df[cols]
    
    # Save to CSV
    r2_df.to_csv(output_file, index=False)
    print(f"R² values saved to '{output_file}'")
    
    return r2_df

def export_summary_stats_csv(data, output_file="model_summary_stats.csv"):
    """
    Calculate and export summary statistics for each model/feature combination
    """
    # Calculate average R² and direction accuracy by model and feature set
    avg_r2 = data.groupby(['model', 'feature_set'])['test_r2'].mean().reset_index()
    avg_dir_acc = data.groupby(['model', 'feature_set'])['test_dir_accuracy'].mean().reset_index()
    
    # Calculate other statistics
    std_r2 = data.groupby(['model', 'feature_set'])['test_r2'].std().reset_index().rename(columns={'test_r2': 'std_r2'})
    std_dir_acc = data.groupby(['model', 'feature_set'])['test_dir_accuracy'].std().reset_index().rename(columns={'test_dir_accuracy': 'std_dir_acc'})
    min_r2 = data.groupby(['model', 'feature_set'])['test_r2'].min().reset_index().rename(columns={'test_r2': 'min_r2'})
    max_r2 = data.groupby(['model', 'feature_set'])['test_r2'].max().reset_index().rename(columns={'test_r2': 'max_r2'})
    min_dir_acc = data.groupby(['model', 'feature_set'])['test_dir_accuracy'].min().reset_index().rename(columns={'test_dir_accuracy': 'min_dir_acc'})
    max_dir_acc = data.groupby(['model', 'feature_set'])['test_dir_accuracy'].max().reset_index().rename(columns={'test_dir_accuracy': 'max_dir_acc'})
    
    # Merge all statistics
    summary_stats = pd.merge(avg_r2, avg_dir_acc, on=['model', 'feature_set'])
    summary_stats = pd.merge(summary_stats, std_r2, on=['model', 'feature_set'])
    summary_stats = pd.merge(summary_stats, std_dir_acc, on=['model', 'feature_set'])
    summary_stats = pd.merge(summary_stats, min_r2, on=['model', 'feature_set'])
    summary_stats = pd.merge(summary_stats, max_r2, on=['model', 'feature_set'])
    summary_stats = pd.merge(summary_stats, min_dir_acc, on=['model', 'feature_set'])
    summary_stats = pd.merge(summary_stats, max_dir_acc, on=['model', 'feature_set'])
    
    # Add combined model_feature column
    summary_stats['model_feature'] = summary_stats['model'] + ' | ' + summary_stats['feature_set']
    
    # Sort rows by model and feature set
    model_order = ['OLS', 'Ridge', 'Lasso', 'KNN', 'RandomForest', 'LSTM']
    feature_order = FEATURE_SETS
    
    # Create a custom sort key
    summary_stats['sort_key'] = summary_stats['model'].map({m: i for i, m in enumerate(model_order)})
    summary_stats['sort_key2'] = summary_stats['feature_set'].map({f: i for i, f in enumerate(feature_order)})
    summary_stats = summary_stats.sort_values(['sort_key', 'sort_key2']).drop(['sort_key', 'sort_key2'], axis=1)
    
    # Rename columns for clarity
    summary_stats = summary_stats.rename(columns={
        'test_r2': 'avg_r2',
        'test_dir_accuracy': 'avg_dir_acc'
    })
    
    # Reorder columns
    cols = ['model', 'feature_set', 'model_feature', 
            'avg_r2', 'std_r2', 'min_r2', 'max_r2',
            'avg_dir_acc', 'std_dir_acc', 'min_dir_acc', 'max_dir_acc']
    summary_stats = summary_stats[cols]
    
    # Save to CSV
    summary_stats.to_csv(output_file, index=False)
    print(f"Summary statistics saved to '{output_file}'")
    
    return summary_stats

def main():
    """
    Main function to load data and export CSV files
    """
    try:
        # Specify the base path to the ticker folders
        base_path = "model_results_v10"
        
        # Process ticker data
        print("Processing ticker data...")
        data, tickers = process_all_ticker_data(base_path)
        
        print(f"Exporting data for {len(tickers)} tickers...")
        
        # Export direction accuracy values to CSV
        export_direction_accuracy_csv(data, tickers)
        
        # Export R² values to CSV
        export_r2_values_csv(data, tickers)
        
        # Export summary statistics to CSV
        export_summary_stats_csv(data)
        
        print("CSV export completed successfully!")
        
    except Exception as e:
        print(f"Error: {str(e)}")
        import traceback
        traceback.print_exc()

if __name__ == "__main__":
    main()
```
